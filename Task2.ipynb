# Build a model using a simple neural network architecture that takes the SMILES calculated fingerprints as input. Train it on the train data and predict the activity of the test set molecules.

```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
from keras.models import Sequential
from keras.layers import Dense
```

```python
train_data = pd.read_csv('train_set_data.csv')
test_data = pd.read_csv('test_set_data.csv')
```

```python
train_data.columns
```

```python
test_data.columns
```

```python
def smiles_to_fingerprint(smiles):
    mol = Chem.MolFromSmiles(smiles)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048)
    fp_array = pd.Series(list(fp.ToBitString()))
    return fp_array

train_fps = train_data['SMILES'].apply(smiles_to_fingerprint)
test_fps = test_data['SMILES'].apply(smiles_to_fingerprint)
```

```python
#X_train = train_fps.values.tolist()
#y_train = train_data['Activity'].values.tolist()
X_train = train_fps.iloc[1:].astype(float).values.tolist()
y_train = train_data['Activity'].iloc[1:].astype(float).values.tolist()

```

```python
X_test = test_fps.iloc[1:].astype(float).values.tolist()
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(2048,)))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
y_pred = model.predict(X_test)
```
